### ✏️ Apply you Inclusion / Exclusion Criteria

### ✏️ 採用/除外基準を適用しましょう

Now that we have done some Initial EDA, let's define criteria for the study cohort selection

初期EDAを行ったので、研究コホートの選択基準を定義してみましょう。

**How to identify Sampling and Representation Bias in a systematic way?**

**サンプリング・バイアス、代表バイアスを体系的に把握する方法とは？**

When we apply inclusion / exclusion criteria, it is important to keep track of the patients that are being left out of the study to spotlight points of sampling bias.

We can create a function that gets the demographic information from a dataframe of patients, to be run this each time we apply an exclusion criteria.

**This is a good practice of study cohort selection!**

組入基準/除外基準を適用する場合、サンプリング・バイアスのポイントを明らかにするために、研究から除外される患者を追跡することが重要です。

患者のデータフレームから人口統計学的情報を取得する関数を作成し、除外基準を適用するたびにこれを実行することができます。

**これは、研究コホート選択の良い練習になります！**

````python
from typing import Dict

def get_demography(df: pd.DataFrame) -> Dict[str, Dict[str, float]]:
    """コホートの人口統計を取得する。

    引数:
        df (pd.DataFrame): コホートのデータフレーム

    返り値:
        dict: 人種、性別、母語などの人口統計学的情報と、それに対応する比率を含む辞書型のデータ
    """
    demo = {}

    demo["race"] = df["race_group"].value_counts(normalize=True).to_dict()

    sex_counts = df["sex_female"].value_counts(normalize=True)
    demo["sex"] = {
        "Male": sex_counts.get(0, 0),
        "Female": sex_counts.get(1, 0)
    }

    demo["language"] = df["language"].value_counts(normalize=True).to_dict()

    # You can add other variables here, following the same structure

    return demo


def print_demo(demo: Dict[str, Dict[str, float]]) -> None:
    """ 出力
      引数:
        dict: 人種、性別、母語などの人口統計学的情報と、それに対応する比率を含む辞書型のデータ
      返り値:
        str: printで出力可能な、人口統計を表示するためのフォーマット付き文字列。
    """

    demo_str = ""
    for key, value in demo.items():
        if isinstance(value, dict):
            demo_str += f"{key}: ["
            for key2, value2 in value.items():
                demo_str += f"{key2}: {round(value2*100,1)}%, "
            demo_str = demo_str[:-2] + "]\n"
        else:
            demo_str += f"{key}: {round(value*100,1)}%, "
            demo_str = demo_str[:-2]

    print(demo_str)
````

````python
# Working Example on the Dataset we just loaded
print_demo(get_demography(df))
````

````python
# To prevent data leakage, we're setting lab results or clinical status with a timedelta of zero or more to NaN.
timedelta_cols = [col for col in df.columns if col.startswith("delta_")]
timedelta_cols_dict = {}
for col in timedelta_cols:
  if col.split("delta_")[-1] in df.columns:
    timedelta_cols_dict[col] = [col.split("delta_")[-1]]
  else:
    print(f"{col}: The corresponding column could not be determined automatically.")
````

````python
# Add corresponding columns manually
timedelta_cols_dict["delta_vent_start"] = ["ventilation_status", "invasive_vent"]
timedelta_cols_dict["delta_vp_start"] = ["norepinephrine_equivalent_dose"]
````

````python
import numpy as np

# Replace with nan
for timedelta_col, corresponding_col in timedelta_cols_dict.items():
  for col in corresponding_col:
    df[[timedelta_col, col]] = df[[timedelta_col, col]].mask(df[timedelta_col]>=0, np.nan)
````

````python
# Code here!
print(f"Original length of data frame: {len(df)}")

df_list = [df]
exclusion_criteria = [
    ("SpO2 is NaN", lambda df: df["SpO2"].isna()),
    ("SaO2 > 100", lambda df: df["SaO2"] > 100),
    ("SpO2 > 100", lambda df: df["SpO2"] > 100),
    ("SaO2 < 65", lambda df: df["SaO2"] < 65),
    ("SpO2 < 65", lambda df: df["SpO2"] < 65),
    ("delta_SpO2 < -10", lambda df: df["delta_SpO2"] < -10)
]

for i, (condition_desc, condition_func) in enumerate(exclusion_criteria):
    print(f"======== {condition_desc} ========")
    condition = condition_func(df_list[-1])
    _df = df_list[-1][~condition].reset_index(drop=True)
    print(f"Removed {len(df_list[-1]) - len(_df)} records with {condition_desc}. The demographics of removed records are as follows.")
    print_demo(get_demography(df_list[-1][condition]))
    print("")
    print(f"As a result, {len(_df)} records were extracted. The demographics of extracted records are as follows.")
    print_demo(get_demography(_df))
    df_list.append(_df)
    print("")

data_final = df_list[-1].copy()
````

### ✏️ Visualizations

### ✏️ 可視化

#### Calibration Maps

#### キャリブレーションマップ

````python
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
from matplotlib.gridspec import GridSpec
from matplotlib import pyplot as plt
import seaborn as sns


def sao2_spo2_plot(df, SaO2_HH=88, SpO2_HH=88,
                   lbl_text="Measured"):

    r2 = r2_score(df.SaO2, df.SpO2)

    fig = plt.figure(figsize=(8,8))
    gs = GridSpec(5,5)

    ax_scatter = fig.add_subplot(gs[1:5, 0:4])

    df_norm = df[~((df.SaO2 < SaO2_HH) & (df.SpO2 >= SpO2_HH))]
    df_hypo = df[(df.SaO2 < SaO2_HH) & (df.SpO2 >= SpO2_HH)]

    hh = len(df_hypo)
    tot = len(df_norm) + len(df_hypo)

    sns.scatterplot(df_norm, x='SpO2', y='SaO2',
                    marker='o', alpha=0.1, ax=ax_scatter)
    sns.scatterplot(df_hypo, x='SpO2', y='SaO2',
                    marker='o', color="tab:red", alpha=0.1, ax=ax_scatter,
                    label="H.H")
    ax_scatter.set_xlabel(f'{lbl_text} SpO2 (%)')
    ax_scatter.set_ylabel('SaO2 (%)')
    ax_scatter.set_xlim([64, 102])
    ax_scatter.set_ylim([64, 102])
    ax_scatter.grid(False)
    ax_scatter.get_legend().remove()

    ax_hist_x = fig.add_subplot(gs[0,0:4])
    sns.kdeplot(x=df.SpO2, ax=ax_hist_x, color='green', fill=True)
    ax_hist_x.xaxis.set_ticklabels([])
    ax_hist_x.grid(False)
    ax_hist_x.set_xlabel('')
    ax_hist_x.set_ylim([0,.35])
    ax_hist_x.axvline(x=df.SpO2.mean(), linewidth=0.8,
                    linestyle='--', color='green',
                    label="SpO2 Mean")


    ax_hist_y = fig.add_subplot(gs[1:5, 4])
    sns.kdeplot(y=df.SaO2, ax=ax_hist_y, color='orange', fill=True)
    ax_hist_y.yaxis.set_ticklabels([])
    ax_hist_y.grid(False)
    ax_hist_y.set_ylabel('')
    ax_hist_y.set_xlim([0,.35])
    ax_hist_y.axhline(y=df.SaO2.mean(), xmin=0, xmax=1,
                    linewidth=0.8, linestyle='--', color='orange',
                    label="SaO2 Mean")

    fig.legend(loc='upper right', bbox_to_anchor=(0.915, 0.85),
               ncol=1, fontsize=9)

    fig.suptitle(f"SaO2 vs SpO2: R\u00B2 = {r2:.2f}, H.H = {hh} / {tot}",
                 fontsize=16, y=0.95)

    plt.show()
````

````python
sao2_spo2_plot(data_final, SaO2_HH=88, SpO2_HH=88, lbl_text="Measured")
````

#### Measurements Distributions

#### 測定値の分布

````python
from typing import List

def sao2_spo2_dists(df: pd.DataFrame, races: List[str]) -> None:
  fig, ax = plt.subplots(len(races), 1, sharex= True, sharey=True,
                        figsize=(2 * len(races), 5))

  for i, r in enumerate(races):
      sns.kdeplot(x=df[df["race_group_binary"] == r]["SaO2"], ax=ax[i],
                  color='orange', fill=True, label="SaO2")
      sns.kdeplot(x=df[df["race_group_binary"] == r]["SpO2"], ax=ax[i],
                  color='green', fill=True, label="SpO2")
      ax[i].set_xlim([90,100])
      ax[i].legend(loc='upper left', ncol=1, fontsize=9)
      ax[i].set_xlabel('O2 Saturation (%)')
      ax[i].set_title(f'{r} Patients')

  fig.suptitle('SaO2 and SpO2 Density Distributions')
  plt.show()
````

````python
# Encode Race Groups
data_final["race_group_binary"] = data_final["race_group"].apply(lambda x: 'Non-White' if x != 'White' else 'White')

races = ["White", "Non-White"]
````

````python
sao2_spo2_dists(data_final, races)
````

#### Gap Distributions

#### 差の分布

````python
def sao2_spo2_gap(df: pd.DataFrame, races: List[str], comparison="White vs. Non-White"):

  fig, ax = plt.subplots(1, 1, sharex= True, sharey=True, figsize=(8,4))

  colors = ['green', 'black']

  for i, r in enumerate(races):
      gap = df[df["race_group_binary"] == r]["SpO2"] - df[df["race_group_binary"] == r]["SaO2"]

      sns.kdeplot(
              x=gap,
              ax=ax,
              color=colors[i],
              fill=True,
              label=r
          )

  ax.set_xlim([-4,4])
  ax.legend(loc='upper left', ncol=1, fontsize=9)
  ax.set_xlabel('SpO2 - SaO2 (%)')
  ax.set_title(f'SpO2 - SaO2 Gap Distribution, {comparison} Patients')

  plt.show()
````

````python
sao2_spo2_gap(data_final, races, comparison="White vs. Non-White")
````

Is it more pronounced across certain groups?

特定のグループ間で顕著なのでしょうか？

````python
# Code here!
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

race_groups = data_final["race_group"].unique()
n = len(race_groups)

# 2列で表示するため、必要な行数を計算します
rows = np.ceil(n/2).astype(int)

# Figureのサイズを調整
fig, axs = plt.subplots(rows, 2, figsize=(12, 5*rows))
axs = axs.ravel() # 1次元配列に変換

for i in range(len(axs)):
    if i < n:
        race = race_groups[i]
        data_subset = data_final[data_final["race_group"] == race]
        counts = data_subset["HH_study"].value_counts()

        sns.barplot(x=counts.index, y=counts.values, ax=axs[i])

        axs[i].set_title(f'HH_study Distribution for {race}', fontsize=14)
        axs[i].set_ylabel('Number of Occurrences', fontsize=12)
        axs[i].set_xlabel('HH_study', fontsize=12)

        total = counts.sum()
        for p in axs[i].patches:
            height = p.get_height()
            # データラベルのフォントサイズを調整
            axs[i].text(p.get_x()+p.get_width()/2.,
                        height + 3,
                        f"{int(height)}, {100 * height/total:.2f}%",
                        ha="center", fontsize=11)
    else:
        # Remove the extra subplots
        fig.delaxes(axs[i])

# サブプロット間のスペースを手動調整
plt.subplots_adjust(wspace=0.2, hspace=0.3)

plt.show()
````
